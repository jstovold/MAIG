```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
totalSamples = int(1e6)

inside = []
outside = []

for i in range(0, totalSamples):
    x = np.random.random()
    y = np.random.random()

    # sqrt(x) < 1  ==  x < 1
    if x**2 + y**2 < 1.0:
        inside.append((x, y))
    else:
        outside.append((x, y))


# inside / total = pi / 4
numInside = float(len(inside))
pi = (numInside / totalSamples) * 4

print(pi)
```

```python
fig = plt.figure()
plt.scatter(np.array(inside)[:,0], np.array(inside)[:,1], c='red', marker='x')#, alpha=0.01)
plt.scatter(np.array(outside)[:,0], np.array(outside)[:,1], c='blue')#, alpha=0.01)
```
